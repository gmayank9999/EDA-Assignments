```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Titanic: Machine Learning from Disaster
# ● Data: Titanic Dataset
# ● Plot the following:
# ○ Histograms: Visualize the age distribution of passengers.
# ○ Box Plot: Compare fare distributions by passenger class.
# ○ Scatter Plot: Age vs. fare to explore correlations.
# ○ Bar Plot: Compare the number of survivors and non-survivors.
# ○ Count Plot: Show the count of passengers in different classes.
```

```python
df=pd.read_csv("titanic.csv")
df
```

```python
#1) Histograms: Visualize the age distribution of passengers.
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), bins=30,kde=True )
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Box Plot: Compare fare distributions by passenger class
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# 3. Scatter Plot: Age vs. Fare to explore correlations
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived')
plt.title('Age vs. Fare (colored by Survival Status)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# 4. Bar Plot: Compare the number of survivors and non-survivors
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Survivors', 'Survivors'])
plt.show()
```

```python
# 5. Count Plot: Show the count of passengers in different classes
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Iris Species
# ● Data: Iris Dataset
# ● Plot the following:
# ○ Pair Plot: Explore relationships between petal and sepal length/width.
# ○ Scatter Plot: Visualize petal length vs. petal width.
# ○ Violin Plot: Compare sepal length distributions by species.
```

```python
df1=pd.read_csv("Iris.csv")
df1
```

```python
# 1. Pair Plot: Explore relationships between petal and sepal length/width
sns.pairplot(df1, hue='Species', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Sepal and Petal Dimensions by Species', y=1.02)
plt.show()
```

```python
# 2. Scatter Plot: Visualize petal length vs. petal width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df1)
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# 3. Violin Plot: Compare sepal length distributions by species
plt.figure(figsize=(8, 6))
sns.violinplot(x='Species', y='SepalLengthCm', data=df1)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# 3. House Prices: Advanced Regression Techniques
# ● Data: House Prices Dataset
# ● Plot the following:
# ○ Histograms: Distribution of house prices.
# ○ Box Plot: Explore price distributions by neighborhood.
# ○ Scatter Plot: Living area vs. sale price.
# ○ Correlation Matrix: Analyze correlations between features like square footage,
# number of rooms, and sale price.
```

```python
df2=pd.read_csv("House Prices - Advanced Regression Techniques.csv")
df2
```

```python
# 1. Histogram: Distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df2['SalePrice'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Box Plot: Explore price distributions by neighborhood
plt.figure(figsize=(12, 8))
sns.boxplot(x='Neighborhood', y='SalePrice', data=df2)
plt.xticks(rotation=90)
plt.title('House Prices by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.show()
```

```python
# 3. Scatter Plot: Living area vs. sale price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df2)
plt.title('Living Area vs. Sale Price')
plt.xlabel('Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.show()
```

```python
# 4. Correlation Matrix: Analyzing correlations between features
# Selecting relevant features for correlation
correlation_features = ['SalePrice', 'GrLivArea', 'TotalBsmtSF', 'GarageCars', 'OverallQual', 'YearBuilt']
correlation_matrix = df2[correlation_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# 4. Student Performance Data
# ● Data: Student Performance Dataset
# ● Plot the following:
# ○ Histograms: Distribution of scores in math, reading, and writing.
# ○ Box Plot: Score distributions by gender or race/ethnicity.
# ○ Bar Plot: Number of students achieving different grade levels.
# ○ Violin Plot: Compare the distribution of scores by gender.
```

```python
df3=pd.read_csv("student-por.csv")
df3
```

```python
# 1.Histograms: Distribution of scores in math, reading, and writing.
# Combined Histogram for G1, G2, and G3 
plt.figure(figsize=(10, 6))
sns.histplot(df3['G1'], bins=10, color='blue', kde=True, label='G1', alpha=0.6)
sns.histplot(df3['G2'], bins=10, color='green', kde=True, label='G2', alpha=0.6)
sns.histplot(df3['G3'], bins=10, color='red', kde=True, label='G3', alpha=0.6)
plt.title("Combined Distribution of G1, G2, and G3 Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()  
plt.show()

```

```python
# 2. Box Plot: Score distributions by gender or race/ethnicity.
# Box plot for score distributions by gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df3, x='sex', y='G3')
plt.title('Score Distribution (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('G3 Score')
plt.show()
```

```python
# 3.Bar Plot: Number of students achieving different grade levels.
bins = [0, 5, 10, 15, 20]
labels = ['Failing', 'Passing', 'Good', 'Excellent']
df3['grade_level'] = pd.cut(df3['G3'], bins=bins, labels=labels)

# Bar plot for number of students by grade level
plt.figure(figsize=(8, 6))
sns.countplot(data=df3, x='grade_level')
plt.title('Number of Students by Grade Level')
plt.xlabel('Grade Level')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 4. Violin Plot: Compare the distribution of scores by gendeer
plt.figure(figsize=(10, 6))
sns.violinplot(data=df3, x='sex', y='G3')
plt.title('Distribution of G3 Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('G3 Score')
plt.show()
```

```python
# 5. Retail Analysis: Walmart Sales Forecasting
# ● Data: Walmart Sales Forecasting
# ● Plot the following:
# ○ Line Plot: Show sales trends over time.
# ○ Heatmap: Correlation of sales with other factors like holiday seasons,
# promotions, etc.
# ○ Scatter Plot: Sales vs. store size or distance to city center.
# ○ Bar Plot: Visualize weekly sales per store.
```

```python
df4=pd.read_csv("WALMART_SALES_DATA.csv")
df4
```

```python
# Line Plot: Show sales trends over time
plt.figure(figsize=(14, 6))
plt.plot(df4.index, df4['Weekly_Sales'], color='blue', linewidth=1)
plt.title("Weekly Sales Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.grid(True)
plt.show()
```

```python
# Heatmap: Correlation of Sales with Other Factors
correlation_matrix = df4[['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Weekly Sales and Other Factors")
plt.show()
```

```python
# Scatter Plot: Sales vs. Store
plt.figure(figsize=(10, 6))
plt.scatter(df4['Store'], df4['Weekly_Sales'], alpha=0.5, color='purple')
plt.title("Sales vs. Store Size/Location")
plt.xlabel("Store ID (proxy for location)")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
# Group by Store and calculate average weekly sales
avg_sales_per_store = df4.groupby('Store')['Weekly_Sales'].mean().sort_values()

# Bar Plot: Visualize weekly sales per store
plt.figure(figsize=(14, 8))
avg_sales_per_store.plot(kind='bar', color='teal')
plt.title("Average Weekly Sales per Store")
plt.xlabel("Store")
plt.ylabel("Average Weekly Sales")
plt.xticks(rotation=90)
plt.show()
```
